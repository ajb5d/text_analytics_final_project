```python
#!/usr/bin/env python3
import pandas as pd
import sqlite3
import plotly.graph_objects as go
from pathlib import Path
```

```python
DATA_DIR = Path("../data")
db = sqlite3.connect("../data/corpus_f4.sqlite")
TERMS = pd.read_sql("SELECT * FROM terms", db)
db.close()
COORDS = pd.read_parquet("../data/corpus_f5_word2vec.parquet")
COORDS = COORDS.merge(TERMS[['term_str', 'count', "tfidf_sum"]], left_on='term', right_on='term_str')
COORDS = COORDS.sort_values('count', ascending=False)
```

# Word Embeddings

```python
fig_data = COORDS.nlargest(250, 'tfidf_sum')

fig = go.Figure(data=go.Scatter(
    x=fig_data['x'],
    y=fig_data['y'],
    mode='text',
    text=fig_data['term'],
))

fig.update_layout(
    width=800,
    height=800,
    template="simple_white",
    title="t-SNE of Word Embeddings for Top 250 Terms by TF-IDF Sum",
    xaxis_title="t-SNE Dimension 1",
    yaxis_title="t-SNE Dimension 2"
)

fig.show()
fig.write_image("embedding_tsne.png")
```
